```python
import os
from google.cloud import storage
```

```python
storage_client = storage.Client(project='soundboard-381415') #set the ID of your project at GCP 

#function for creating the bucket at the google cloud services 
def create_bucket(bucket_name):
    bucket = storage_client.create_bucket(bucket_name)
```

```python
source_folder = "../data/musdb18"

# Replace with the name of the GCP bucket you want to upload the files to
bucket_name = "soundboard_data_musdb"

# Instantiate a client object
client = storage.Client()

# Get a reference to the destination bucket
bucket = client.get_bucket(bucket_name)

# Iterate over the files in the source folder
for root, dirs, files in os.walk(source_folder):
    for file in files:
        # Get the full path to the file
        file_path = os.path.join(root, file)

        # Get the path of the file relative to the source folder
        relative_path = os.path.relpath(file_path, source_folder)

        # Get the path to the folder containing the file in the GCP bucket
        destination_folder = os.path.dirname(relative_path)

        # Get a reference to the destination blob in the bucket
        blob = bucket.blob(destination_folder + "/" + file)

        # Upload the file to the bucket
        blob.upload_from_filename(file_path)

        print(f"Uploaded {file_path} to {destination_folder}")
```

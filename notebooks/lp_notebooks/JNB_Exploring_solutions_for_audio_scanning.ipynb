# Scanning function that returns the loudest second of a video

## Basic logic flow

1. From a YouTube URL:
    1. Pulls down the audio-video data using YouTube from pytube
    2. Seperates into audio and visual using download function via YouTube
    3. Convert into mp3 from mp4 using pydub
    3. Takes the audio and scans through for the highest peaks (ranked, top 5)
    4. With peak as midpoint (0.5 of the second), take total volume over 1s period for all top 5s
    5. Return timestamp of loudest second
    6. Play video second of returned timestamp

2. Packages used:
    1. **Librosa** - converting audio into plot for analysis
    2. **PyTube** - for capturing YT content with audio
    3. **PyDub** - for converting into mp3
    4. **iPython.Display** - music player within JNB
    5. **os, glob** - renaming filenames while handling directory paths
